# Yield Curve Discriptive Statistics

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv('yeildcurve.csv')
df['Date']= pd.to_datetime(df['Date'])
```

```python
fullrecdf= pd.read_csv('us.csv')
fullrecdf['DATE']= pd.to_datetime(fullrecdf['DATE'])
fullrecdf['JHDUSRGDPBR']= fullrecdf['JHDUSRGDPBR'].astype(bool)
```

```python
values = [1.59,1.56,1.57,1.59,1.54,1.61,1.61,1.65,1.75,1.84,2.16,2.31]
names = ['1 Mo','2 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
plt.xlabel('Bond Duration')
plt.ylabel('Interest Rate (%)')
plt.title('Interest Rates on 15 November 2019')
plt.plot(names,values)
plt.scatter(names,values)
plt.show()
```

```python
values = [5.02,5.14,5.18,5.1,4.94,4.85,4.81,4.81,4.81,5.0,4.9]
names = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
plt.xlabel('Bond Duration')
plt.ylabel('Interest Rate (%)')
plt.title('Interest Rates on 23 January 2007')
plt.plot(names,values)
plt.scatter(names,values)
plt.show()
```

```python
recdf = fullrecdf[(fullrecdf['DATE'].dt.year >= 1990)]
plt.plot(recdf['DATE'],recdf['JHDUSRGDPBR'],color="red")
plt.xlabel('Date')
plt.ylabel('Recession')
plt.title('Recessions 1990 onwards')
plt.show()
```

```python
y=df['10 Yr']
x=df['Date']
plt.xlabel('Date')
plt.ylabel('Interest Rate (%)')
plt.plot(x,y)
plt.title("10 Year Treasury Bond vs. Time")
plt.show()
```

```python
#10-1
newdf = df[['Date','1 Yr','10 Yr']]
newdf['sub'] = newdf['10 Yr']- newdf['1 Yr']
newdf.info()
y=newdf['sub']
x=newdf['Date']
plt.xlabel('Date')
plt.ylabel('Difference in Interest Rate (%)')
plt.title("10 Year Yield - 1 Year Yield vs. Time")
plt.plot(x,y)
plt.axhline(y=0.0, color='black', linestyle='-')
plt.show()
```
