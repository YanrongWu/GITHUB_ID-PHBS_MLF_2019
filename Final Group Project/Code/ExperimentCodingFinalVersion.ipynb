```python
#############################################
#######   Final Project by Group 12    ######
#############################################

#basic library
import pandas as pd
import numpy as np

# Prepare for the train set and test set
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Functions for Generation
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

# Features Selection
from sklearn.decomposition import PCA

# Algorithm for ML
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Functions for evaluationb
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import learning_curve

# Functions to visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for i in range(1,28):
    
    inputPath = 'DATASET_FOR_CASES\DATASET_1_27_MULTIPLE\DATASET_'+ str(i) +'.csv'
    df= pd.read_csv(inputPath)
    print('CASE_'+str(i)+'#Dataframe:\n')
    print(df)
    X, y = df.values[:, 1:-1], df.values[:, -1]
    Project_X_Train, Project_X_Test, Project_y_Train, Project_y_Test =\
        train_test_split(X, y, 
                         test_size=0.3, 
                         random_state=0, 
                         stratify=y)
    Project_y_Train = Project_y_Train.astype('int')
    Project_y_Test = Project_y_Test.astype('int') 

    lr = OneVsRestClassifier(LogisticRegression(random_state=1))
    svm = OneVsRestClassifier(SVC(random_state=1))
    tree = OneVsRestClassifier(DecisionTreeClassifier(criterion='gini', random_state=1))
        
    pipe_lr = make_pipeline(StandardScaler(), PCA(n_components=1), lr)
    pipe_svm = make_pipeline(StandardScaler(),PCA(n_components=1), svm)
    pipe_tree = make_pipeline(StandardScaler(), PCA(n_components=1), tree)
    
    
    pipe_lr.fit(Project_X_Train, Project_y_Train)
    pipe_svm.fit(Project_X_Train, Project_y_Train)
    
    
    
    param_grid_LR = [{'onevsrestclassifier__estimator__C': [0.1, 1.0, 10]}]
    param_grid_SVM = [{'onevsrestclassifier__estimator__C': [0.1, 1.0, 10], 'onevsrestclassifier__estimator__kernel': ['linear']}]
    param_grid_DT = [{'onevsrestclassifier__estimator__max_depth': [1,2,3,4,5,6,7,8,9]}]
    # param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]    
    # param_grid_SVM = [
    #           {'onevsrestclassifier__estimator__C': param_range, 
    #            'onevsrestclassifier__estimator__gamma': param_range,
    #            'onevsrestclassifier__estimator__kernel': ['rbf']}]
    
    gs_LR = GridSearchCV(estimator=pipe_lr, param_grid=param_grid_LR, scoring='accuracy', cv=2)
    gs_SVM = GridSearchCV(estimator=pipe_svm, param_grid=param_grid_SVM, scoring='accuracy', cv=2)
    gs_DT = GridSearchCV(estimator=pipe_tree, param_grid=param_grid_DT, scoring='accuracy', cv=2)

    scores_LR = cross_val_score(gs_LR, Project_X_Train, Project_y_Train, scoring='accuracy', cv=10, n_jobs=4)
    print('CASE_'+str(i)+'#'+'scores_LR: CV accuracy: %.3f +/- %.3f' % (np.mean(scores_LR),np.std(scores_LR)))

    scores_SVM = cross_val_score(gs_SVM, Project_X_Train, Project_y_Train, scoring='accuracy', cv=10, n_jobs=4)
    print('CASE_'+str(i)+'#'+'scores_SVM: CV accuracy: %.3f +/- %.3f' % (np.mean(scores_SVM), np.std(scores_SVM)))

    scores_DT = cross_val_score(gs_DT, Project_X_Train, Project_y_Train, scoring='accuracy', cv=10, n_jobs=4)
    print('CASE_'+str(i)+'#'+'scores_DT: CV accuracy: %.3f +/- %.3f' % (np.mean(scores_DT), np.std(scores_DT)))

   
    
    gs_LR = gs_LR.fit(Project_X_Train, Project_y_Train)
    print('CASE_'+str(i)+'#'+'LR Best Score: '+str(gs_LR.best_score_))
    print('CASE_'+str(i)+'#'+'LR Best Params: '+str(gs_LR.best_params_))
    
    gs_SVM = gs_SVM.fit(Project_X_Train, Project_y_Train)
    print('CASE_'+str(i)+'#'+'SVM Best Score: '+str(gs_SVM.best_score_))
    print('CASE_'+str(i)+'#'+'SVM Best Params: '+str(gs_SVM.best_params_))

    gs_DT = gs_DT.fit(Project_X_Train, Project_y_Train)
    print('CASE_'+str(i)+'#'+'DT Best Score: '+str(gs_DT.best_score_))
    print('CASE_'+str(i)+'#'+'DT Best Params: '+str(gs_DT.best_params_))
    
    
    plt.clf()
    print('LR painting... ...')
    train_sizes, train_scores, test_scores =\
                learning_curve(estimator=gs_LR.best_estimator_,
                               X=Project_X_Train,
                               y=Project_y_Train,
                               train_sizes=np.linspace(0.1, 1.0,10),
                               cv=10,
                               n_jobs=1)
    
    title = 'Data: DATASET_'+str(i)
    title += '__'
    title += 'Model:LR'
    title += '__'
    title += 'Best Paras: '
    title += str(gs_LR.best_params_)
    title += '__'
    title += 'Best Scores: '
    title += str(gs_LR.best_score_)
    title += '__'
    title += 'CV Accuracy: '
    title += str(np.mean(scores_LR))
    title += '+-'
    title += str(np.std(scores_LR))
    
    s = 'Data: DATASET_'+str(i)
    s += '\n'
    s += 'Model:LR'
    s += '\n'
    s += 'Best Paras: '
    s += str(gs_LR.best_params_)
    s += '\n'
    s += 'Best Scores: '
    s += str(gs_LR.best_score_)
    s += '\n'
    s += 'CV Accuracy: '
    s += str(np.mean(scores_LR))
    s += '+-'
    s += str(np.std(scores_LR))
    print(s)
    
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.plot(train_sizes, train_mean,
             color='blue', marker='o',
             markersize=5, label='training accuracy')

    plt.fill_between(train_sizes,
                     train_mean + train_std,
                     train_mean - train_std,
                     alpha=0.15, color='blue')

    plt.plot(train_sizes, test_mean,
             color='green', linestyle='--',
             marker='s', markersize=5,
             label='validation accuracy')

    plt.fill_between(train_sizes,
                     test_mean + test_std,
                     test_mean - test_std,
                     alpha=0.15, color='green')

    plt.grid()
    plt.xlabel('Number of training samples')
    plt.ylabel('Accuracy')
    plt.legend(loc='upper right')
    maxValue = max(np.append(train_scores,test_scores))
    minValue = min(np.append(train_scores,test_scores))
    plt.ylim([minValue,maxValue])
    plt.text(0, -0.4, s, fontdict=None, fontsize = 'medium', horizontalalignment = 'left')
#     plt.tight_layout()
    title = "LR_CASE_"+str(i)
    plt.savefig('RESULT_IMAGES/'+title+'.png')
#     plt.show()


    plt.clf()
    print('SVM painting... ...')
    train_sizes, train_scores, test_scores =\
            learning_curve(estimator=gs_SVM.best_estimator_,
                           X=Project_X_Train,
                           y=Project_y_Train,
                           train_sizes=np.linspace(0.1, 1.0,10),
                           cv=10,
                           n_jobs=1)
    
    title = 'Data: DATASET_'+str(i)
    title += '__'
    title += 'Model:SVM'
    title += '__'
    title += 'Best Paras: '
    title += str(gs_SVM.best_params_)
    title += '__'
    title += 'Best Scores: '
    title += str(gs_SVM.best_score_)
    title += '__'
    title += 'CV Accuracy: '
    title += str(np.mean(scores_SVM))
    title += '+/-'
    title += str(np.std(scores_SVM))
    
    s = 'Data: DATASET_'+str(i)
    s += '\n'
    s += 'Model:LR'
    s += '\n'
    s += 'Best Paras: '
    s += str(gs_SVM.best_params_)
    s += '\n'
    s += 'Best Scores: '
    s += str(gs_SVM.best_score_)
    s += '\n'
    s += 'CV Accuracy: '
    s += str(np.mean(scores_SVM))
    s += '+/-'
    s += str(np.std(scores_SVM))
    print(s)
    
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.plot(train_sizes, train_mean,
             color='blue', marker='o',
             markersize=5, label='training accuracy')

    plt.fill_between(train_sizes,
                     train_mean + train_std,
                     train_mean - train_std,
                     alpha=0.15, color='blue')

    plt.plot(train_sizes, test_mean,
             color='green', linestyle='--',
             marker='s', markersize=5,
             label='validation accuracy')

    plt.fill_between(train_sizes,
                     test_mean + test_std,
                     test_mean - test_std,
                     alpha=0.15, color='green')

    plt.grid()
    plt.xlabel('Number of training samples')
    plt.ylabel('Accuracy')
    plt.legend(loc='upper right')
    maxValue = max(np.append(train_scores,test_scores))
    minValue = min(np.append(train_scores,test_scores))
    plt.ylim([minValue,maxValue])
    plt.text(0, -0.4, s, fontdict=None, fontsize = 'medium', horizontalalignment = 'left')
#     plt.tight_layout()
    title = "SVM_CASE_"+str(i)
    plt.savefig('RESULT_IMAGES/'+title+'.png')
#     plt.show()

    plt.clf()
    print('DT painting... ...')
    train_sizes, train_scores, test_scores =\
            learning_curve(estimator=gs_DT.best_estimator_,
                           X=Project_X_Train,
                           y=Project_y_Train,
                           train_sizes=np.linspace(0.1, 1.0,10),
                           cv=10,
                           n_jobs=1)
    
    title = 'Data: DATASET_'+str(i)
    title += '__'
    title += 'Model:LR'
    title += '__'
    title += 'Best Paras: '
    title += str(gs_DT.best_params_)
    title += '__'
    title += 'Best Scores: '
    title += str(gs_DT.best_score_)
    title += '__'
    title += 'CV Accuracy: '
    title += str(np.mean(scores_DT))
    title += '+/-'
    title += str(np.std(scores_DT))
    
    s = 'Data: DATASET_'+str(i)
    s += '\n'
    s += 'Model:LR'
    s += '\n'
    s += 'Best Paras: '
    s += str(gs_DT.best_params_)
    s += '\n'
    s += 'Best Scores: '
    s += str(gs_DT.best_score_)
    s += '\n'
    s += 'CV Accuracy: '
    s += str(np.mean(scores_DT))
    s += '+/-'
    s += str(np.std(scores_DT))
    print(s)
    
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.plot(train_sizes, train_mean,
             color='blue', marker='o',
             markersize=5, label='training accuracy')

    plt.fill_between(train_sizes,
                     train_mean + train_std,
                     train_mean - train_std,
                     alpha=0.15, color='blue')

    plt.plot(train_sizes, test_mean,
             color='green', linestyle='--',
             marker='s', markersize=5,
             label='validation accuracy')

    plt.fill_between(train_sizes,
                     test_mean + test_std,
                     test_mean - test_std,
                     alpha=0.15, color='green')

    plt.grid()
    plt.xlabel('Number of training samples')
    plt.ylabel('Accuracy')
    plt.legend(loc='upper right')
    maxValue = max(np.append(train_scores,test_scores))
    minValue = min(np.append(train_scores,test_scores))
    plt.ylim([minValue,maxValue])
    plt.text(0, -0.4, s, fontdict=None, fontsize = 'medium', horizontalalignment = 'left')
#     plt.tight_layout()
    title = "DT_CASE_"+str(i)
    plt.savefig('RESULT_IMAGES/'+title+'.png')
#     plt.show()

    print("CASE_"+str(i)+"_FINISHED ########################################################################")
```
